```python
free_count=paid_count=free_sum=paid_sum=0
for i in range(len(df.values)):
    if(df.values[i][4]==0 and df.values[i][11]=='Sports'):
        free_count=free_count+1
        free_sum=free_sum+df.values[i][7]
    else:
        paid_count=paid_count+1
        paid_sum=paid_sum+df.values[i][7]
#print(free_count,free_sum,paid_count,paid_sum)
avg_free=free_sum/free_count
avg_paid=paid_sum/paid_count
print(avg_free)
print(avg_paid)

#y=2
#x=[avg_free,avg_paid]
#idth=1/1.5
#lt.bar(x,y,width,color='blue')
#lt.show()
```

```python
## Average Rating of free vs Paid

import numpy as np
#x=len(str((y))
x=[avg_free,avg_paid]
y=np.arange(1,len(x)+1)
width=1/10
plt.bar(x,y,width,color='blue')
plt.xlabel('Free                                                                     Paid')
plt.title('Avg rating of Sports category in Free vs Paid Apps')
plt.show()
```
