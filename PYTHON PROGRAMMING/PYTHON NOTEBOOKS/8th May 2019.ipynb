```
#  1)Testing Divisibility with 2 and 3 but not 4

def checkdivisibility(n):
  if(n%2==0 and n%3==0 and n%4!=0):
    return True
  else:
    return False
  

  
checkdivisibility(6)
checkdivisibility(16)
checkdivisibility(20)
```

```
#   2)Given 2 ints, a and b, return True if one if them is 10 or if their sum is 10.
def check(a,b):
  if(a==10 or b==10 or (a+b)==10):
    return True
  else:
    return False
  
  
check(10,9)
check(5,5)
check(9,9)
```

```
#     3) Given an int n, return True if it is within 10 of 100 or 200. Note: abs(num) computes the absolute value of a number.
def check(n):
  if(abs(100-n)<=10 or abs(200-n)<=10):
     return True
  else :
     return False
     
     
     
check(90)     
check(93)
check(89)
check(105)
check(115)
```

```
#     4) Finding the greatest of 3 numbers
def great(x,y,z):
  if(x > y and x > z):
    return x 
  elif(x>y and z>x):
    return z  
  elif(x<y and y>z):
    return y  
  elif(x<y and y<z):
    return z 
  
  
  
  
great(3,4,5)  
```

```
#    5)  Determining whether a number is Even or Odd
def oe(n):
  if (n%2==0):
    return 'even'
  else :
    return 'odd'
  
  
  
  
oe(4) 
```
