```python
import numpy as np 

t=np.arange(0.0,2.0,0.1)
v=1+np.sin(2*np.pi*t)
plt.xlabel('time')
plt.ylabel('voltage')
plt.plot(t,v)
plt.grid(True)
plt.title('UFO inputs')
plt.savefig('FIGURES/UFO Inputs')
plt.show()
```

### Problem 2: 
#### Problem Statement : 
Plot a line graph on in income vs year for the state of Arizona

```python
import pandas as pd
df = pd.read_csv('DataFiles/income.csv')
year=[]
income=[]
for i in range(2,11):
    year.append(df.columns[i])
for i in range(2,11):
    income.append(df.values[2][i])
year
income
plt.xlabel("years")
plt.ylabel("income")
plt.plot(year,income)
plt.grid(True)
plt.title("INCOME IN  Arizona")
plt.savefig('FIGURES/INCOME OF Arizona IN LINEGRAPH')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('DataFiles/marks.csv')
marks=[]
for i in range(len(df.values)):
    marks.append(df.values[i][1])
marks
maxi=max(marks)
mini=min(marks)
avg=sum(marks)/len(marks)
dc=0
fc=0
sc=0
tc=0
fa=0
for i in range(1,len(marks)):
    if(marks[i]>=80): dc=dc+1
    if(marks[i]>=60 and marks[i]<80): fc=fc+1
    if(marks[i]>=50 and marks[i]<60): sc=sc+1
    if(marks[i]>=40 and marks[i]<50): tc=tc+1
    if(40>marks[i]): fa=fa+1
lables=['DIST','FIRST','SECOND','THIRD','FAIL']
sizes=[dc,fc,sc,tc,fa]
explode=(0.1,0,0,0,0)
plt.pie(sizes,explode=explode,labels=lables,startangle=90)
plt.title("REPORT OF 1ST YEAR ")
plt.show()
```
